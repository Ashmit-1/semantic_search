### Cleaning the dataset.
The dataset is messy. Apart from the actual content it contains email like headers, footers spaces etc. In order to make the vector embeddings proper I got rid of most of unnecessary information. 
#### What I have removed:
- Headers(Newsgroups, Path, From, Message-ID, Sender, Organization, References, Date, Lines, etc.)
- Footers: separated with "--" 2 hyphens in most of the document where they are present.
- Empty spaces: There are many unnecessary empty spaces here and there in the document. I removed them.

#### What I have kept:
- Subject: Most of the articles have a subject, I have kept them as they convey relevant information. 
- The actual text content between the header and footer. 

```python
import pandas as pd
df = pd.read_csv("newsgroups_raw.csv")
```

```python
import re

def clean_document(text):

    subject = ""

    # extract subject line with regex if it is present
    subject_match = re.search(r"Subject:\s*(.*)", text)
    if subject_match:
        subject = subject_match.group(1).strip()

    #remove rest of the header
    parts = text.split("\n\n", 1)
    text = parts[1] if len(parts) > 1 else parts[0]

    #remove footer
    #remove only when 2 hyphens are encountered because there are many instances
    #where multiple hyphens are present denoting something else and not the footer
    text = re.split(r'(?<!-)\n--(?!-)', text)[0]

    #remove whitespace
    text = re.sub(r"\s+", " ", text)

    text = text.strip()

    # combine subject and body
    if subject:
        text = f"{subject}. {text}"

    return text
```

```python
df["clean_text"] = df["text"].apply(clean_document)
```
