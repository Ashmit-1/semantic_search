### Convet to vector embeddings
The cleaned data is now converted to vector embeddings. 
Here I have used all-MiniLM-L6-v2 as it is a perfect fit as a  fast, memory efficient embedding model. 
<br>
There were other embedding models but some were heavy, for some the performance (embedding quality) was not good. 
There was the option of choosing an LLM embedding model but that was firstly not fully free of cost. And I think for our application that is not needed. 
<br>
I have exported the embeddings as a numpy array for further use and also stored them as csv .

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

def generate_embeddings(texts, batch_size=64):

    embeddings = model.encode(
        texts,
        batch_size=batch_size,
        show_progress_bar=True
    )

    return embeddings
```

```python
import pandas as pd

df = pd.read_csv("newsgroups_cleaned.csv")
texts = df["clean_text"].tolist()

embeddings = generate_embeddings(texts)

print("Embedding shape:", embeddings.shape)
```

```python
import numpy as np

#convert to list and then save as numpy array as well as csv
df["embedding"] = embeddings.tolist()
np.save("embeddings.npy", embeddings)
df.to_csv("vectorized_data.csv", index=False)
```
